**Bloco 5 — describe()**
> A função `describe` traz uma série de análises estatísticas que ajudam a entender a distribuição dos dados, como a contagem de valores preenchidos, a média, o desvio padrão e os valores mínimo e máximo de cada coluna.


```python
print("Verificar valores negativos em order_value:", (df['order_value'] < 0).sum())
print("Verificar valores negativos em delivery_delay_days:", (df['delivery_delay_days'] < 0).sum())
print("Verificar NPS fora do intervalo de 0 a 10:", ((df['nps_score'] < 0) | (df['nps_score'] > 10)).sum())
print("Idade minima:", df['customer_age'].min(), "| Idade maxima:", df['customer_age'].max())
```

**Bloco 6 — Verificação de consistência**
> Aqui verifico se algumas colunas específicas apresentam valores que não deveriam ser possíveis, como valores negativos ou notas de NPS fora do intervalo de 0 a 10.


```python
print(df['customer_region'].value_counts())
```

**Bloco 7 — Verificação de região**
> Aqui verifico se a discrepância entre as regiões da base de dados.


```python
def classificar_nps(score):
    if score >= 9:
        return 'Promotor'
    elif score >= 7:
        return 'Neutro'
    else:
        return 'Detrator'

df['nps_category'] = df['nps_score'].apply(classificar_nps)
print(df['nps_category'].value_counts())
```

**Bloco 8 — Classificação do NPS**
> Aqui crio uma função para classificar o NPS conforme a metodologia oficial: notas maiores ou iguais a 9 são Promotores, entre 7 e 8 são Neutros, e as demais (0 a 6) são Detratores. Em seguida aplico a função na base e observo que 84% dos clientes são Detratores — um sinal claro de um problema grave de satisfação a ser investigado na análise exploratória.


```python
df.to_csv("../data/processed/nps_tratado.csv", index=False)
print("Arquivo nps_tratado.csv salvo com sucesso na pasta data/processed.")
```
